## What is OpenZeppelin? 

- It's a library for writing secure Smart Contracts on the Ethereum Blockchain.
- You can use it to build decentralized applications, protocols and organizations
- It's a library written in the Solidity Programming Language
- Solidity is the Ethereum Virtual Machine compatible language for writing smart contracts
- Solidity compiles down to EVM bytecode, thats why a more popular language isn't well-suited for it
- More popular programming languages like C++ and Python have constructs better suited for comping down to regular bytecode
- Its meant to provide secure, tested and community-audited code

## Setup Steps

- OpenZeppelin integrates with Truffle, an Ethereum development environment. 
- Install NPM if you haven't from here https://www.npmjs.com/get-npm 
- To Install Truffle using NPM and initialize your project, enter the following commands in terminal

```javascript
npm install -g truffle
mkdir myproject && cd myproject
truffle init
```

- To install the OpenZeppelin library, run the following in your Solidity project root directory:

```javascript
npm init -y
npm install -E openzeppelin-solidity
```

- OpenZeppelin does not currently follow semantic versioning. 
- You may encounter breaking changes upon a minor version bump. 
- Pin the version of OpenZeppelin you use, as done by the -E (--save-exact) option.
- After that, you'll get all the library's contracts in the node_modules/openzeppelin-solidity/contracts folder. 
- You can use the contracts in the library like so:

```javascript
import 'openzeppelin-solidity/contracts/ownership/Ownable.sol';

contract MyContract is Ownable {
  ...
}
```
